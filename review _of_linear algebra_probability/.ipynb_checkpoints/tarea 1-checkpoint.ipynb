## Singular Value Decomposition (SVD)


SVD es un método estable y eficaz para dividir el sistema en un conjunto de componentes linealmente independientes, cada uno de los cuales tiene su propia contribución de energía.


De tal forma que el sistema se pude representar como un matriz $M$ donde esa matriz que es rectangular $mxn$  existe una descomposición de valor singular de  $M$, de la forma <a href="http://ethen8181.github.io/machine-learning/dim_reduct/svd.html">[1]</a><a href="https://medium.com/data-science-group-iitr/singular-value-decomposition-elucidated-e97005fb82fa">[2]</a><a href="https://medium.com/@rameshputalapattu/jupyter-python-image->compression-and-svd-an-interactive-exploration-703c953e44f6">[3]</a> $$M=U\Sigma V^T$$

- $U$ - es una matriz unitaria $mxr$ ortogonal. (vector singular izquierdo)
- $\Sigma$ - matriz diagonal con valores singulares $rxr$
- $V$ - es una matriz unitaria $nxr$ ortogonal , (vector singular derecho)

$\Sigma$ (Sigma) es una matriz diagonal de orden decreciente no negativa de $rxr$. Todos los elementos que no están en la diagonal principal son 0 y los elementos de $\Sigma$ se denominan valores singulares. Otra notación común que se utiliza para esta matriz es $S$. <a href="http://ethen8181.github.io/machine-learning/dim_reduct/svd.html">[1]</a>


### Matriz ortogonal
se refiere a una matriz cuadrada donde las columnas están a 90 grados entre sí y su producto punto interno es cero, es decir, dada una matriz ortogonal$A$ , $AA^T=AA^T=I$ y $A^T=A^{-1}$ <a href="http://ethen8181.github.io/machine-learning/dim_reduct/svd.html">[1]</a>


## Interpretación de SVD
### Interpretación geométrica
La expresión $U\Sigma V^T$se puede interpretar como una composición de tres transformaciones geométricas: una rotación, una escala y otra rotación

veamos un ejemplo: 

se tiene una matriz $X = \begin{bmatrix}
    -10 & -10 & 20 & 20\\
    -10 & 20 & 20 &-10\\\end{bmatrix}$ y una matriz de transformación $A = \begin{bmatrix}
    -1 & 0.3 \\
    0.45 & 1.2 \\\end{bmatrix}$

para la imagen mostrada a  continuación, donde el cuadrado punteado muestra $X$  y la matriz transformada $Ax$ como la forma solida.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.preprocessing import normalize, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer


def plot_points(matrix, ls='--', lw=1.2, colors=None):

  x_points, y_points = matrix
  size = len(x_points)
  colors = ['red', 'blue', 'orange', 'green'] if not None else colors
  for i in range(size):
    plt.plot(x_points[i], y_points[i], color=colors[i], marker='o')
    plt.plot([x_points[i], x_points[(i+1) % size]], 
             [y_points[i], y_points[(i+1) % size]], 
             color=colors[i], linestyle=ls, linewidth=lw)
    
def plot_point_label(prefix, point, location):
 
  plt.annotate('{0}={1}'.format(prefix, tuple(map(lambda x: round(x, 2), point))), 
  xy=point, xytext=location, textcoords='data', color='white', weight='semibold',
  bbox=dict(fc='navy', alpha=0.6, ec='none'), arrowprops=dict(arrowstyle='->'))
    
    
matrix=np.array([[-10, -10, 20 ,20], [-10, 20, 20 ,-10]])
A = np.matrix([[1, 0.3], [0.45, 1.2]])
transformed_matrix = A * matrix

plot_points(matrix)
plot_points(transformed_matrix.A, '-', lw=3.0)
plt.annotate('X', xy=(20.2, 5),textcoords='data', weight='semibold')
plt.annotate('Ax', xy=(10, 28),textcoords='data', weight='semibold')


ax = plt.axes() 
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')



```

```python

U, S, V = np.linalg.svd(A)
print('Matriz unitaria (vector singular izquierdo):')
print(U)
print('matriz diagonal con valores singulares :')
print(S)      
print('Matriz unitaria (vector singular derecho):')
print(V )
```

$V^Tx$ Podemos ver que la multiplicación por $V^T$gira y refleja la matriz de entrada $x$. Observe el intercambio de colores rojo-azul y verde-amarillo que indica un reflejo a lo largo del eje $x$


```python
V= np.matrix([[-0.63586997, -0.77179621], [-0.77179621,  0.63586997 ]])

matrix_VT= V * matrix

plot_points(matrix)
plot_points(matrix_VT.A, '-', lw=3.0)
plt.annotate('X', xy=(20.2, 5),textcoords='data', weight='semibold')
plt.annotate('V x', xy=(-20.2, 5),textcoords='data', weight='semibold')

# Set the axes
ax = plt.axes() 
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')



```

$SV^Tx$ Dado que $S$ solo contiene valores en la diagonal, escala la matriz. $V$ gira la matriz a una posición donde los valores singulares ahora representan el factor de escala a lo largo de la base $V$. En la siguiente imagen, $V^Tx$ es la grafica punteada y $SV^Tx$ es sólido.




```python
S= np.matrix([[1.49065822, 0], [0,  0.71444949 ]])

matrix_SVT=  S *V * matrix


plot_points(matrix_VT.A)
plot_points(matrix_SVT.A, '-', lw=3.0)
plt.annotate('V x', xy=(-8, 20),textcoords='data', weight='semibold')
plt.annotate('SV x', xy=(-35, 6),textcoords='data', weight='semibold')

# Set the axes
ax = plt.axes() 
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
```

$USV^Tx$ Por último,  $U$ Gira y refleja la matriz de nuevo a la base estándar.


```python
U= np.matrix([[-0.58189652, -0.81326284], [-0.81326284  ,0.58189652 ]])

matrix_USVT= U* S *V * matrix


plot_points(matrix_SVT.A)
plot_points(matrix_USVT.A, '-', lw=3.0)
plt.annotate('SV x', xy=(-8, 22),textcoords='data', weight='semibold')
plt.annotate('USV x', xy=(-35, 6),textcoords='data', weight='semibold')

# Set the axes
ax = plt.axes() 
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
```

## Descomposicion en valores propios  (eigens ) 

Recapitulación  un poco el tema de algebra lineal  , una matriz $x$ puede transformarse multiplicándola con una matriz cuadrada $nxn$, $A$. Así, la matriz transformada puede ser representada por la ecuación $nxn$ $$Ax=b$$

antes de que entremos en detalle que son las descomposición de valores propios tenesmos que saber que es un vector propio y valor propio 

hay un vector en partícula lo veremos a continuación con la grafica ,pero pensemos un momento en la aplicación lineal que vimos anteriormente $$X = \begin{bmatrix} -10 & -10 & 20 & 20\\ 
-10 & 20 & 20 &-10\\\end{bmatrix}$$

imaginémonos que en ese sistema líneal ,hay  un vector en particular y su sistema generador de ese vector pasa una  lineal atreves de su origen y su punta, si quisiéramos mirara que paso con ese vector en particular cuando se hace la transformación, muy probablemente no esté en la misma lineal , en el sistema generador que creamos imaginativamente ,ya que la mayoría de vectores se van a ver desplazados respecto a su sistema generador durante la transformación, pero algunos vectores especiales si se mantienen en su propio sistema generador a estos vectores se les llama vectores propios y cada vector propio tiene asociado un valor propio que no es más que el factor por el cual el vector se alarga o  encoge durante la transformación

$$Ax=\lambda x$$

Donde $A$representa una aplicacion lineal 
$x$ se llama un vector propio que, cuando se multiplica con $A$, produce un valor escalar, $\lambda$, denominado valor propi.

masomenos lo que que dice aqui mas formalmente
Cualquier vector $v$ en la línea realizada desde los puntos que pasan por el origen $(0,0)$  y un vector propio son todos vectores propios. Transformar $v$ al multiplicarlo por la matriz de transformación $A$ o su valor propio asociado $\lambda$ resultará en el mismo vector

Eigenvalue $\lambda$ nos dice cuánto $x$ se escala, se estira, se encoge, se invierte o no se toca cuando se multiplica por A. El número de valores propios es, como máximo, el número de dimensiones,$N$. Por lo tanto, un conjunto de vectores $2D$ tendrá como máximo 2 valores propios y vectores propios correspondientes


```python
plot_points(matrix)
plot_points(transformed_matrix.A, '-', lw=3.0)
plt.annotate('X', xy=(20.2, 5),textcoords='data', weight='semibold')
plt.annotate('Ax', xy=(10, 28),textcoords='data', weight='semibold')


ax = plt.axes() 
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

evals, evecs = np.linalg.eig(A)
print('evals');print(evals);print('evecs');print(evecs)

x_v1, y_v1 = evecs[:,0].getA1()
x_v2, y_v2 = evecs[:,1].getA1()

m1 = y_v1/x_v1 # Gradient of 1st eigenvector
m2 = y_v2/x_v2 # Gradient of 2nd eigenvector

p1 = [-10/m1, -10]  # 1st point at y = -10
p2 = [20/m2, 20]    # 2nd point at y = -20
trans_p1 = A*np.matrix(p1).T
trans_p2 = A*np.matrix(p2).T

# Plot eigenvectors and labels
plt.plot([x_v1*-50, x_v1*50], [y_v1*-50, y_v1*50], color='royalblue')
plt.plot([x_v2*-50, x_v2*50], [y_v2*-50, y_v2*50], color='crimson')
plt.annotate('e1', xy=(-20, 14),textcoords='data', weight='semibold', color='royalblue')
plt.annotate('e2', xy=(-11, -20),textcoords='data', weight='semibold', color='crimson')

```

## Numero de Condicion de una Matriz


En esta sección vamos a ver cuándo una matriz se encuentra bien condicionada o mal condicionada, este indicador se asocia con la estabilidad de su solución para el  sistema que se quiere resolver, en este caso se  tiene una ecuación línea de la forma $Ax  =  b $ .Supongamos que se hace un pequeño cambio en la matriz  $A$ ese pequeño cambio también se debe ver reflejado en la mismas proporciones en $b$. Por lo tanto, lo que se quiere es entender cuan estable es la solución con respecto a la perturbación de los datos. 


<center>Sistema a solucionar :  $ Ax  =  b $ </center>  


Teniendo en cuenta lo anterior ¿Donde se añade la perturbación?
Se podría pensar, si yo quiero ver cómo responde el sistema a ciertas entra, entonces perturbo $b$ sumado un termino $r$ ,por lo tanto la solución o la consecuencia  va estar dada por $\bar{x}$. 


<center> perturbación :  $ A\bar{x}  =  b+r$ </center>  

Ahora bien miremos esa diferencia de error relativo dado en esas dos diferencias para mirar a que  velocidad   la solución, x , cambiará con respecto a un cambio en b

Donde el error relativo 

sea $x$ la solución de $Ax = b$, y $\bar{x}$ la solución calculada de $A\bar{x} = b+r$, por lo que

- El error en la solución es $e = x - \bar{x} $  

- error relativo en la solución es la distancia entre las dos soluciones y un factor de normalización que es la división entre la norma del valor verdadero $x$ .  
$$e_{x}=\dfrac {\parallel e \parallel} {\parallel x \parallel}=\dfrac {\parallel  x - \bar{x} \parallel} {\parallel x \parallel}$$

En general, no podemos verificar el error o el error relativo, por lo que una posible forma de comprobarlo es comprobando la perturbación en el sistema $Ax = b$

- residual $r$ es la diferencia 

$$A\bar{x} - b = r$$
$$b= Ax$$
$$r = Ax - A\bar{x}$$ 


- error residual relativo 


$$ e_{b}=\dfrac {\parallel Ax - A\bar{x} \parallel} {\parallel Ax\parallel}=\dfrac {\parallel r  \parallel} {\parallel b \parallel}$$


Ahora comprendamos esta la relación de solución para este sistema

$$r = Ax - A\bar{x}$$ 
$$r = A(x - \bar{x})$$   

$$ A^{- 1} r =x - \bar{x} $$ 
Teniendo en cuenta esta ecuación podemos rescribir el error relativo con respeto a $x$
$$e_{x}=\dfrac {\parallel  x - \bar{x} \parallel} {\parallel x \parallel}= \dfrac {\parallel  A^{- 1} r \parallel} {\parallel x \parallel}\leq \dfrac {\| A^{- 1}\| \parallel   r \parallel}  {\parallel x \parallel}$$

A continuación se manipula la alguna expresión para obtener un estimado en términos de $\dfrac {  1 }  {\parallel x \parallel} $

 $$Ax = b \Rightarrow \parallel Ax\parallel = \parallel b\parallel \Rightarrow \parallel  A \parallel \parallel  x\parallel \geq  \parallel b\parallel \Rightarrow   \dfrac {  1 }  {\parallel x \parallel}  \leq \dfrac { \parallel A \parallel}  {\parallel b \parallel}  $$
 
relaciona una vez se tenga el termino queda nuevamente

 $$e_{x} \leq \dfrac {\| A^{- 1}\| \cdot \parallel   r \parallel}  {\parallel x \parallel}  \leq \| A^{- 1}\| \cdot \parallel   r \parallel \dfrac {\parallel  A \parallel} {\parallel b \parallel}={\| A\| \parallel    A^{- 1} \parallel}  e_{b}=k(A).e_{b} $$
 
$$ k(A)=\lVert A^{- 1} \rVert \lVert A \rVert \geq \lVert A^{- 1} A \rVert =1$$ numero de condición de A


usando $ l_{2}-norma$


$$ k(A)=\lVert A^{- 1} \rVert_{2} \lVert A \rVert _{2}=\dfrac{ max_{i}\| \lambda_{i}(A)\|}{ min_{i}\| \lambda_{i}(A)\|}$$

Cuando $A $no es simétrico, el máximo de $ \dfrac { \parallel Ax \parallel } { \parallel x \parallel} $ se puede encontrar en un vector $x$ que no es uno de los vectores propios. Este máximo es una excelente medida del tamaño de $A$. 

La norma de $A$ es el número $\parallel A \parallel$ definido por


$$\parallel A \parallel=max_{x\neq 0} \dfrac { \parallel Ax \parallel } { \parallel x \parallel}$$


La norma de $A$ mide la mayor cantidad por la cual cualquier vector (vector propio o no) se amplifica mediante la multiplicación de la matriz $ \dfrac { \parallel Ax \parallel } { \parallel x \parallel}$ 

La norma de la matriz de identidad es 1. Para calcular la norma, cuadrar ambos lados para alcanzar el $A A^T $ simétrico

$$\lVert A \rVert _{2}=max \dfrac { \parallel Ax \parallel } { \parallel x \parallel}=max \dfrac { x^T A^T A x} { x^Tx }=\lambda_{max}(A^T A)$$

$$\lVert A^{- 1} \rVert _{2}=min \dfrac { \parallel Ax \parallel } { \parallel x \parallel}=min \dfrac { x^T A^T A x} { x^Tx }=\lambda_{min}(A^T A)$$


Cuando el número de condición es exactamente uno (que solo puede suceder si A es un múltiplo escalar de una isometría lineal ), entonces un algoritmo de solución puede encontrar (en principio, lo que significa que si el algoritmo no introduce errores propios) una aproximación de la solución cuya precisión no es peor que la de los datos.

Sin embargo, no significa que el algoritmo converja rápidamente a esta solución, solo que no divergirá arbitrariamente debido a la inexactitud de los datos de origen (error hacia atrás), siempre que el error de avance introducido por el algoritmo no divida también. Por acumular errores de redondeo intermedios.

El número de condición también puede ser infinito, pero esto implica que el problema está mal planteado (no posee una solución única y bien definida para cada elección de datos; es decir, la matriz no es invertible), y ningún algoritmo puede ser Se espera encontrar de forma fiable una solución.

La definición del número de condición depende de la elección de la norma, como puede ilustrarse con dos ejemplos.
El número de condición mide la sensibilidad de la salida de una función a su entrada. Cuando dos variables predictoras están altamente correlacionadas, lo que se denomina multicolinealidad, los coeficientes o factores de esas variables predictoras pueden fluctuar erráticamente para pequeños cambios en los datos, o el modelo. Idealmente, los modelos similares deberían ser similares, es decir, tener coeficientes aproximadamente iguales. La multicolinealidad puede causar que la inversión de la matriz numérica se dañe o produzca resultados inexactos. 

Calculamos el número de condición tomando los valores propios del producto de las variables predictoras (incluido el vector constante de unos) y luego tomando la raíz cuadrada de la relación del valor propio más grande al valor propio menor. Si el número de condición es mayor que treinta, entonces la regresión puede tener multicolinealidad.<a href="https://www.youtube.com/watch?v=PZ9cHKTdEXM&t=4s">[9]</a> 




```python
from numpy import linalg as LA


cond = LA.cond(A)
print('Condición l2 largest sing value No.: {:.5f}'.format(cond))

#frobenius
condfro = LA.cond(A, 'fro')
print('Condición por Norma de frobenius No.: {:.5f}'.format( condfro ))


#max(sum(abs(x), axis=1))
condmaxinf = LA.cond(A, np.inf)
print('Condición por Norma max infinito No.: {:.5f}'.format( condmaxinf ))

#min(sum(abs(x), axis=1))
condmininf = LA.cond(A, np.inf)
print('Condición por Norma min infinito No.: {:.5f}'.format( condmininf ))



#max(sum(abs(x), axis=0))
condmaxuno = LA.cond(A, 1)
print('Condición por Norma max l1 No.: {:.5f}'.format( condmaxuno ))

#min(sum(abs(x), axis=0))
condminuno = LA.cond(A, -1)
print('Condición por Norma min l1 No.: {:.5f}'.format( condminuno ))


#2-norm (largest sing. value)
conddos = LA.cond(A, 2)
print('Condición por Norma l2 largest sing value No.: {:.5f}'.format( conddos ))

#smallest singular value
conddos = LA.cond(A, -2)
print('Condición por Norma l2 smallest singular value No.: {:.5f}'.format( conddos ))

#2-norm (largest sing. value)
A= np.matrix( A )
EV = np.linalg.eig( A*A.T)
CN = np.sqrt( EV[0].max() / EV[0].min() )
print('Condition l2 largest sing value ecuacion No.: {:.5f}'.format( CN ))

#smallest singular value
smalsv = min(LA.svd(A,compute_uv=0))*min(LA.svd(LA.inv(A), compute_uv=0))
print('Condition l2 largest sing value ecuacion No.: {:.5f}'.format( smalsv  ))


```

# Rango de una matriz
En los términos más simples, el rango de una matriz es literalmente el "rango" de ella.El quid de esta definición es esencialmente  el Rango (otra palabra para espacio de columna) . Sea una matriz $A$ que es $m × n$, con vectores de columna $v_{1} ,  v_{2}, ...,  v_{n}$ . Una combinación lineal de estos vectores es cualquier vector de la forma.
$C_{1}v_{1} ,C_{2 }v_{2}, ..., C_{n}v_{n}$
donde $C{1},C_{2 } , ..., C_{n}$ son escalares. El conjunto de todas las posibles combinaciones lineales de $v_{1} ,  v_{2}, ...,  v_{n}$ se llama el espacio de la columna de $A$ . Es decir, el espacio de columnas de $A$ es el intervalo de los vectores $v_{1} ,  v_{2}, ...,  v_{n}$ .

Cualquier combinación lineal de los vectores de columna de una matriz A se puede escribir como el producto de A con un vector de columna:

$$A\begin{bmatrix} C_{1}   \\ \vdots \\
C_{n} \\\end{bmatrix}=\begin{bmatrix} 
    a_{11} & \dots  & a_{1n} \\
    \vdots & \ddots & \vdots\\
    a_{m1} & \dots & a_{mn} 
    \end{bmatrix}\begin{bmatrix} C_{1}   \\ \vdots \\
C_{n} \\\end{bmatrix}=\begin{bmatrix} 
    C_{1}a_{11}+ & \dots  & +C_{n}a_{1n} \\
    \vdots & \ddots & \vdots\\
     C_{1}a_{m1}+ & \dots &+ C_{n}a_{mn} 
    \end{bmatrix}=C_{1}\begin{bmatrix} a_{11}  \\ \vdots \\
a_{m1} \\\end{bmatrix}+\dots+C_{n} \begin{bmatrix} a_{1n}  \\ \vdots \\
a_{mn }\\\end{bmatrix}$$
