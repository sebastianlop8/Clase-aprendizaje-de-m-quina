## Gradient Descent 

En regresión lineal, el modelo se enfoca en obtener la línea de regresión de mejor ajuste para predecir el valor de $y$ en función del valor de entrada dado $(x)$.Mientras se entrena el modelo, el modelo calcula la función de costo que mide el error cuadrático medio entre el valor predicho (pred) y el valor verdadero (y). La pérdida es el error en nuestro valor predicho de $\theta_1$ y $\theta_2$ . Nuestro objetivo es minimizar este error para obtener el valor más preciso de $\theta_1$ y $\theta_2$. 

Muchas de las funciones objetivas en el aprendizaje automático son difíciles de minimizar analíticamente, pero podemos aproximarnos al mínimo utilizando el descenso de gradiente.
$$ J(w,b) = \frac{1}{m} \sum_{i=0}^m (\bar y_i - y_i)^2 $$ 
$$ J(w,b) = \frac{1}{m} \sum_{i=0}^m ((wx_i-b) - y_i)^2 $$  
$$ \frac{d}{dw}J(w,b) = \frac{d}{dw}(\frac{1}{m} \sum_{i=0}^m ((wx_i-b) - y_i)^2)=-\frac{2}{m}\sum_{i=0}^m x_i((wx_i-b) -  y_i)$$ 
$$ \frac{d}{db}J(w,b) = \frac{d}{db}(\frac{1}{m} \sum_{i=0}^m ((wx_i-b) - y_i)^2)=-\frac{2}{m}\sum_{i=0}^m 
((w x_i-b) -  y_i)$$


```python
import numpy as np
import matplotlib.pyplot as plt

t0 = 4         # Parametro theta 0
t1 = 3        # Parametro theta 1
ruido = 2
X = ruido * np.random.rand(100, 1)
Y = t0 + X*t1 + np.random.randn(100, 1)

plt.plot(X, Y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])

```

```python
# inicialización de para metros
w = 1  # inicialización parametro theta 0
d = 3  # inicialización parametro thet1a

L = 0.1     # La tasa de aprendizaje
epochs = 10 # El número de iteraciones para realizar pendiente de gradiente

m = float(len(X)) # Número de elementos en X

# Realización de Pendiente de Gradiente

for i in range(epochs): 
    Y_pred = w*X + d  # El valor actual predicho de Y
    D_w = (-2/m) * sum(X * (Y - Y_pred))  # Derivado w
    D_d = (-2/m) * sum(Y - Y_pred)  # Derivado d
    w = w - L * D_w  # Actualizar w
    d = d - L * D_d  # Actualizar b
    style = "b-" if i > 0 else "r--"         # not shown
    plt.plot(X, Y_pred, style)        # not shown
print (w,d)

plt.plot(X, Y, "b.")                                 # not shown
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.xlabel("$x_1$", fontsize=18)                     # not shown
plt.ylabel("$y$", rotation=0, fontsize=18)           # not shown
plt.axis([0, 2, 0, 15])                              # not shown
plt.show()                                           # not shown
```

```python

import numpy as np
import numpy.random as rnd

np.random.seed(42)
```

```python
t0 = 0.5         # Parametro theta 0
t1 = 2       # Parametro theta 1
t2= 1       # Parametro theta 1
ruido = 6
model_order = 3
m = 100
X = ruido * np.random.rand(m, 1) - 3
Y_S = t0 * X**2 +t2 *X + t1 
Y = t0 * X**2 +t2 *X + t1 + np.random.randn(m, 1)

plt.plot(X, Y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
plt.show()
```

```python
t0 = 0.1         # Parametro theta 0
t1 = 1.5       # Parametro theta 1
t2= 0.2      # Parametro theta 1
L = 0.1     # La tasa de aprendizaje
epochs = 10 # El número de iteraciones para realizar pendiente de gradiente

for i in range(epochs): 
    Y_pred = t0 * X**2 +t2*X + t1     # El valor actual predicho de Y
   
    t0g = (-2/m) * sum(Y - Y_pred) 
    t1g = (-2/m) * sum(X * (Y - Y_pred)) 
    t2g = (-2/m) * sum(X * (Y - Y_pred)) 
    
    t0 = t0 - L * t0g  # Actualizar w
    t1 = t1 - L * t1g  # Actualizar w
    t2 = t2 - L * t2g  # Actualizar w
```

```python
y_model = np.polyval([t0,t2,t1], np.linspace(-4,10,100))

plt.plot(X, Y , "b.")                                 # not shown
plt.plot(np.linspace(-4,10,100),y_model ) # predicted
plt.xlabel("$x_1$", fontsize=18)                     # not shown
plt.ylabel("$y$", rotation=0, fontsize=18)           # not shown
plt.axis([-3, 3, 0, 10])                             # not shown
plt.show() 

```
